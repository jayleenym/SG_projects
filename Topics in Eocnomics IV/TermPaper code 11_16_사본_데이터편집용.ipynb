1. 개인, 기관, 외국인 투자자 비율 변화
  - 개인 투자자에 집중
2. ATR지수 
   <!-- $$avg(max(max(h-l,abs(c(1)-h)),abs(c(1)-l)),Period)$$ -->

   $$avg(max(h_t - l_{t}, |c_{t-1} - h_t|, |c_{t-1} - l_t|), Period)$$
  - TR: (당일고가 - 당일저가), (당일고가 - 전일종가) , (전일종가 - 당일저가) 중에 최대값
  - Period : 주로 14일
